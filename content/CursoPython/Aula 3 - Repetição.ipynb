### Estruturas de Repetição
- Permitem que um bloco de comandos seja executado diversas vezes
- Dois tipos de Repetição:
* Repetição condicional: executa um bloco de código enquanto uma condição lógica for verdadeira (while)
*  Repetição contável: executa um bloco de código um número predeterminado de vezes (for)

### Repetição condicional Enquanto



### Executa o bloco de instruções enquanto a condição for verdadeira
* A condição é uma expressão booleana que pode fazer uso de quaisquer operadores
*  O bloco de código pode conter um ou mais comandos
* O início e o fim do bloco são definidos de acordo com a endentação
* A estrutura de repetição é chamada de loop porque continua-se voltando ao início da instrução até que a condição se torne falsa
* Deve haver algum processo dentro do bloco de comandos que torne a condição falsa para que a repetição seja encerrada
* Quando a condição se torna falsa, a próxima instrução após o bloco do while é executada
* Se a condição do while for falsa desde o início, o bloco de instruções nunca é executado

Exemplo: 
- Você quer arrecadar R$ 500,00 para o casamento do seu amigo.




```python
total = 0 
while (total<500):
    nome=input("Entre com o nome: ")
    valor=float(input("Entre com a contribuição:"))
    print(nome, "contribuiu com", valor)
    total+=valor
print("Objetivo alcançado",total)
```

## Repetição condicional Repita-até
- Não existe estrutura correspondente ao repita no Python
- Caso necessário pode-se fazer estruturas equivalent



- Exemplo
- Você quer arrecadar R$ 500,00 para o casamento do seu amigo.


```python
total = 0 
while (True):
    nome=input("Entre com o nome: ")
    valor=float(input("Entre com a contribuição:"))
    print(nome, "contribuiu com", valor)
    total+=valor
    if(total>=500.0):
        break
print("Objetivo alcançado",total)
```

## Repetição contável
- Executa o bloco de instruções utilizando uma variável de controle
- O laço começa com a inicialização da variável de controle com o primeiro valor.
- A cada iteração a variável de controle é atualizado até atingir o valor final desejado.
- O bloco de código pode conter um ou mais comandos
- O início e o fim do bloco são definidos de acordo com a endentação
- A estrutura do Python oferece algumas versatilidades

```python
# O programa a seguir imprimirá Hello dez vezes
for i in range(10):
    print('Hello')
```

- A estrutura de um loop for é a seguinte:



- O programa abaixo pede ao usuário um número e imprime seu quadrado, então pede
outro número e imprime seu quadrado, etc. 
- Ele faz isso três vezes e depois imprime que o loop é feito.

```python
for i in range(3):
    num = eval(input('Entre com um numero: '))
    print ('O Quadrado do numero é', num*num)
print('O loop foi encerrado.')
```

## A variável de loop
- Há uma parte de um loop for que é um pouco complicada, que é a variável do loop. no exemplo
abaixo, a variável do loop é a variável i. 
- A saída deste programa serão os números 0, 1, . . . ,99, cada um impresso em sua própria linha

```python
for i in range(100):
    print(i)
```

## A função **range**
- O valor que colocamos na função *range* determina quantas vezes faremos um loop.
- O comando produz uma lista de números de zero ao valor menos um. 
- Por exemplo, *range(5)* produz cinco valores: 0, 1, 2, 3 e 4
- Se quisermos que a lista de valores comece em um valor diferente de 0, podemos fazer isso especificando o valor inicial. A instrução *range(1,5)* produzirá a lista 1, 2, 3, 4.

```python
for i in range(1,5):
    print(i)
```

- Outra coisa que podemos fazer é aumentar o incremento em mais de um por vez. 
- Para fazer isso vamos especificar um terceiro argumento. 
- *range(1,10,2)* gera uma lista de dois em dois, produzindo 1, 3, 5, 7, 9.

```python
for i in range(1,10,2):
    print(i)
```

- Para obter a lista de valores invertida, podemos usar um passo de -1. 
- Por exemplo, *range(5,1,-1)* produzirá os valores 5, 4, 3, 2, nessa ordem. 

```python
for i in range(5,1,-1):
    print(i)
```
