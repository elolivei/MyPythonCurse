## OPERADORES ARITMÉTICOS

Para realizar operações matemáticas, podemos utilizar os operadores aritméticos

### operadores aritimeticos 
**'+'** Soma

**'-'** Subtração
 
**'*'** Multiplicação

**'/'** Divisão

**'%'** Devolve o resto da divisão

```python
import random
a = random.randint(-10,10)
b = random.randint(-10,10)
# operadores aritimeticos
# + - * / %
resultado = a % b
print("Operando 1:" , a)
print("Operando 2:" , b)
print("Resultado " , resultado)
```

###  Operadores aritimeticos e atribuicao
Já que acabamos de tratar dos operadores aritméticos, agora apresentaremos os operadores de atribuição, 
desde aquele mais simples e direto 
* como o sinal de igual (=) que, nós, professores, gostamos de chamar de “recebe” 
* até as abreviações envolvendo operadores aritméticos e de atribuição.		

```python
import random
a = random.randrange(1,52)
b = random.randrange(1,52)
# operadores aritimeticos e atribuicao
# += -= *= /= %=
print("Operando a:" , a)
print("Operando b:" , b)
a-=b    # a = a + 10
print("Operando após subtrair a-b " , a)
print("Operando a:" , a)
print("Operando b:" , b)
```

###  Operadores incremento e decremento
**+=1** Incrementa

**-=1** Decrementa

```python
import random
a = random.randrange(1,52)
b = random.randrange(1,52)
# operadores incremento e decremento
# ++ --
print("Operando Antes a= " , a)
print("Operando Antes b= " , b)
#incrementa
a+=1
#decrementa
b-=1

print("Operando Depois a= " , a)
print("Operando Depois b= " , b)
```

### Operador unário de negação
**not**   Negação lógica

```python
		
verdade =  True
# operador de negacao
# !
print("Operando " , verdade)
mentira = not verdade
print("Operando negado " , mentira)
```

###  Operadores de comparação

Como o nome já diz, esses Operadores são usados para comparar dois valores:

```python
import random
a = random.random() * 10.0
b = random.random() * 10.0
# operadores de comparacao
# > >= < <= == != 
print("Operando a= " , a)
print("Operando b= " , b)
verdade = a <= b
print("Comparando " , verdade)
```

### Operadores logicos
	
* Esses Operadores nos possibilitam construir um tipo de teste muito útil e muito utilizado em qualquer programa Python: os testes lógicos.

* Python nos disponibiliza três tipos de Operadores Lógicos: o and, o or e o not.
		

```python
# operadores logicos
print("Operador and")
#tabela da verdade
print("True  and True:  " , (True and True))
print("True  and False: " , (True and False))
print("False and True:  " , (False and True))
print("False and False: " , (False and False))
```

## ENTRADA DE DADOS E CONVERSÃO DE TIPOS
* Até então, vínhamos declarando uma variável atribuindo valores a elas. 
* No entanto, em um cenário real, geralmente o usuário interage com o programa informando dados de entrada. * Em Python, utiliza-se a função **input()**

```python
aluno = input("Digite seu nome: ")
periodo_corrente = input("Digite seu período corrente: ")
ira = input("Digite seu IRA(Índice de Rendimento Acadêmico): ")
laurea = input("Informe se você é o aluno laureado(a) da turma: ")
print(type(aluno))
print(type(periodo_corrente))
print(type(ira))
print(type(laurea))
```

* É importante destacar que a função input() sempre retornará o valor recebido no formato de uma string,  que ele seja de um tipo de dado diferente como, por exemplo, um número real.
* com strings não é possivel usar funções aritiméticas, então o que fazer?
* Basta usar as funções de conversão int(), float() e bool() para conversão.

```python
aluno = input("Digite seu nome: ")
periodo_corrente = int(input("Digite seu período corrente: "))
ira = float(input("Digite seu IRA(Índice de Rend. Acadêmico): "))
laurea = bool(input("Informe se você foi laureado(a) na turma: "))
print(type(aluno))
print(type(periodo_corrente))
print(type(ira))
print(type(laurea))
```
