### Vetores
* Variável composta unidimensional
* Contém espaço para armazenar diversos valores
* É acessada via um índice
* Um vetor nada mais é que uma lista

```python
Vetor = ["Maça","Banana","Abacate"]
for i in Vetor:
    print(i)
Vetor[1] = "Laranja"
for i in Vetor:
    print(i)
```

### Matrizes
* Variável composta multidimensional
* É equivalente a um vetor, contudo permite a utilização de diversas dimensões acessadas via diferentes índices
* Pode ser pensada como um vetor onde cada célula é outro vetor,recursivamente
* Em diversas situações matrizes são necessárias para correlacionar informações
* Matrizes nada mais são que Listas alinhadas

### Listas alinhadas
Uma lista aninhada consiste na existência de uma lista como sendo elemento de uma outra lista.<br> 
**familia = ["João Nascimento", ["João Vitor", "Maria Clara"]]**
* As listas aninhadas podem ser utilizadas para a representação de matrizes.

* Para acessar um elemento da lista aninhada A (ou simplesmente matriz), é preciso informar o índice correspondente e o índice da lista interna. 
* A sua sintaxe é: **lista [\<indice_externo\>][\<indice_interno\>]**

```python

A = [[2, 1, -5], [3, 7, 0], [6, 4, 8]]

tr_A = A[0][0] + A[1][1] + A[2][2]
print(f"tr(A) = {tr_A}")

# Soma dos elementos
soma_A = A[0][0] + A[0][1] + A[0][2]
soma_A += A[1][0] + A[1][1] + A[1][2]
soma_A += A[2][0] + A[2][1] + A[2][2]

print(f"soma(A) = {soma_A}")
```

### Exemplos
* Programa que cria uma matriz n x m preenchida com zeros

```python
def criaMatriz ( linhas=1, colunas=1):

    # cria uma linha com x colunas
    linha_com_zeros = [0]*colunas
    matriz=[]

    for i in range(linhas):
        matriz.append(linha_com_zeros.copy())

    #retorna uma matriz com x linhas
    return (matriz)

matriz = criaMatriz(4,5) 
print(matriz)
matriz[1][3] = 33
print(matriz)
```

### Dimensões da matriz
* É possível usar len() para saber a dimensão de uma matriz
* Assumindo que todas as linhas possuem o mesmo número de elementos, pode-se fazer:
    * Número de linhas:len(matriz)
    * Número de colunas:len(matriz[0])


```python
matriz = []
parar = 0
while parar < 3:
    linha = [parar] * 10
    matriz.append(linha.copy())
    parar += 1
print(matriz)
print("A matriz possui %d linhas" % len(matriz))
print("A matriz possui %d colunas" % len(matriz[0]))
```

### Python permite misturar tipos em uma matriz
* Exemplo: programa que armazena os nomes e idades de pessoas em uma matriz, e imprime o nome da pessoa mais


```python
Matriz = [["Ana",10],["Lucas",5],["Bia",13],["Bruno",8]]
#procura a pessoa mais nova

posicaoMenor = 0
# avança linha a linha
for i in range (len (Matriz)):
    if Matriz[i][1] < Matriz[posicaoMenor][1]:
        posicaoMenor = i
print("A Pessoa mais nova na lista é",Matriz[posicaoMenor][0],"sua idade é",
       Matriz[posicaoMenor][1])
```

Você pode usar : (<inicio:fim>) no indice para indicar um range e fazer um split no vetor. O valor indicado no fim, não é considerado.
No exemplo abaixo somamos os elemento da posição 5 e 6 do vetor na linha 0 da matriz.
*soma = sum(matriz[0][5:7])*


```python
from random import randint
#Operação = str(input())
 
linha=[]
matriz=[]
 
# captura as linhas
for j in range(4):
    # captura uma linha
    for i in range(12):
        x = randint(-100,100)
        linha.append(x)
    matriz.append(linha.copy())
    linha.clear()
print(matriz)
print(matriz[0][5:7])
soma = sum(matriz[0][5:7])
print (soma)
```
