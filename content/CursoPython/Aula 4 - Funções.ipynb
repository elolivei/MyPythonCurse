### O que é uma função
* Nome para um conjunto de comandos que pode ser chamado várias vezes em pontos diferentes do programa, sem a necessidade de repetição de código. 
* o Mais interessante disso tudo é que já utilizamos várias funções em nossos programas Python. 
* Lembre-se de int(),de float(), de input(), de print(), de max(), de range() etc.

### DECLARANDO FUNÇÕES
* Para declarar uma função em Python, precisamos conhecer quatro coisas: a palavra-chave def, o nome da função, a lista de parâmetros e o corpo (bloco de comandos). 



```python
def exibirMensagem():
    print("Python é fácil")
exibirMensagem()
exibirMensagem()
exibirMensagem()
```

### Ordem Declaração
* Python é uma linguagem interpretada, ou seja, a execução dos comandos ocorre sequencialmente, obedecendo, basicamente, a ordem e disposição das linhas de comando. 
Isso significa dizer que não se pode chamar uma função antes de sua declaração. 

```python
# Este programa gera erro
# A chamada foi feita antes de declarar a função
exibirMensagemBoasVindas()

def exibirMensagemBoasVindas():
    print("Seja bem vindo")
```

### Função com parâmetros
* Podemos chamar a função por posição dos parametros ou por nomeação

```python
# Função declarada com dois parâmetros
def exibirMensagemBoasVindas(pessoa, mensagem):
    print(f"{mensagem}, {pessoa}")

exibirMensagemBoasVindas("João","Oi")  # Chamada de função posicional

# Chamando com parametros nomeados
exibirMensagemBoasVindas(pessoa = "Ana", mensagem ="hello")

# com parametros nomeados a ordem não importa
exibirMensagemBoasVindas( mensagem ="hi", pessoa = "Jonh")
```

###  declarar funções com parâmetros contendo valores padrão. 
* Isso significa dizer que agora você pode até ignorar a passagem de parâmetros

```python
# Função declarada com dois parâmetros contendo valores padrão
def exibirMensagemBoasVindas(pessoa = "Fulano", mensagem = "oi"):
    print(f"{mensagem}, {pessoa}")

exibirMensagemBoasVindas()  # Os parâmetros são opcionais
```
