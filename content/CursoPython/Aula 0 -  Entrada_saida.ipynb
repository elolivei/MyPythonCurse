## Entrada e Saída de Dados no Terminal do Python

* Quando tratamos de Programação, é comum imaginar que queremos trocar dados com o computador para resolvermos nossos problemas, certo? Essa troca de dados, quando ocorre diretamente, muitas vezes é chamada Entrada/Saída (ou mais conhecida como I/O, do inglês, Input/Output).
* Quando programamos em Python, usamos principalmente o terminal para ler as informações que a ferramenta nos entrega (como dados, erros e avisos) e o resultado das operações que estamos realizando.



## Entrada de dados com a Função input



- Em Python, fazemos isso utilizando a função input(), que é literalmente ‘entrada’ em inglês.

- A função input() recebe como parâmetro uma string que será mostrada como auxílio ao usuário, geralmente o informando que tipo de dado o programa está aguardando receber.

```python
# Entrando com um String
nome = input("Digite seu nome: ")
print (nome)
```

- O exemplo acima funciona para obter texto do usuário. 
- Para obter números do usuário para usar em cálculos, precisamos fazer algo extra.

```python
# Obter a entrada como inteiro
idade = int(input('Digite sua idade: '))
print(idade)
#Obter a entrada como float
altura = float(input('Digite sua altura: '))
print(altura)
# Obter a entrada como bolean
casado = bool(input('Você é casado: '))
print(casado)
```

- A função **eval** converte o texto digitado pelo usuário em um número. Uma característica legal disso é
você pode inserir expressões, como 3*12+5, e eval irá computá-las para você.

```python
num = eval(input('Entre com um numero: '))
print('O quadrado do seu numero: ', num*num)
```

## Saída de dados com a função Print



* Um dos primeiros contatos que temos com Python é a função print(). 
* Usamos essa função para mostrar algo no terminal do Python. 
* Para concatenar (juntar) informações textuais com dados armazenados nas nossas variáveis, podemos **interpolar** essas variáveis, e até expressões, com nossa String usando o caractere **" , **.

```python
print("olá")
```

- A função de impressão requer parênteses em torno de seus argumentos. 
- No programa acima, é apenas argumento é a string 'Olá'. 
- Qualquer coisa entre aspas será (com algumas exceções) impressa exatamente como aparece. 

```python
#string
print('3+4')
#valor
print(3+4)
```

- Para imprimir várias coisas de uma só vez, separe-as por vírgulas. 
- Python irá inserir espaços automaticamente entre eles. 
- Abaixo está um exemplo e a saída que ele produz.

```python
print('The value of 3+4 is', 3+4)
print('A', 1, 'XYZ', 2)
```

## Marcador de posição

Forma de formatação de strings mais antiga na linguagem. 
Atualmente, se recomenda utilizar as outras formas apresentadas mais adiante no artigo. 
No entanto, como ainda há muito código legado por aí, é importante saber como utilizar.

Como o nome sugere, utiliza-se de marcadores de posição, como %d e %s, no interior da string. 

```python
nome="Elias"
idade=45

# Formatação 
print("Nome: %s Idade: %d" % (nome ,idade)) 
```

* Lidando com Numeros

```python
num = -13.744

print('Números como %s são convertidos para string' % num)
print('Também podem ser representados como inteiros: %d; repare que sem arredondamento' % num)
print('Mostrando o sinal em inteiros: %+d' % num)
print('Mostrando o sinal em floats: %+f' % num)
```

* Observe acima que a representação de float ficou com mais decimais que o número passado. Podemos modificar esta representação com x.yf, onde x é o mínimo de caracteres e y, as casas decimais:

```python

num = 13.788
print('Mínimo de 5 caracteres com 2 casas decimais: %5.2f' % num)
print('Mínimo de 1 caracteres com 0 casas decimais: %1.0f (observe o arredondamento)' % num)
print('Mínimo de 1 caracteres com 5 casas decimais: %1.5f' % num)
print('Mínimo de 10 caracteres com 2 casas decimais: %10.2f' % num)
print('Mínimo de 10 caracteres com 2 casas decimais: %+10.2f (com sinal)' % num)
```

## Método format

Forma introduzida no Python 3.0 e que possui mais flexibilidade e legibilidade que a maneira apresentada anteriormente.

```python
print('Vai entrar um texto {}.'.format('aqui'))
print('Controlando {2} {1}{0}'.format('!', 'ordem', 'a'))
print('Primeiro: {a}; Segundo: {b}; Terceiro: {c}'.format(a=1, b='2', c=12.3))
print('A variável {p} pode ser reutilizada: {p}.'.format(p='(oi!)'))
```

# Lidando com numeros

```python
num = 13.788

print('10 caracteres e 2 casas decimais:{:10.2f}'.format(num))
```
